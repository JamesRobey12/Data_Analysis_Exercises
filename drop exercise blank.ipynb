### pandas.drop() cheats

##### 3 main arguments

1. A singular string or list of strings representing the names of the columns or rows you wish to drop. (Note: When dropping rows, you specify the int index, not a string) 
2. **axis =** This means the axis to select what to drop from. Axis 0 is rows and axis 1 is columns.
3. **inpalce =** This means do you want the changes to affect the current dataframe or return a new dataframe where the changes have been made. By default it is set to False, meaning it will return a new dataframe.
